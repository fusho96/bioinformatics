```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(font_scale=1.5)
```

```python
data = pd.read_csv('taxa.csv',index_col=0)
sns.clustermap(data, z_score=1, cmap="RdYlBu_r",row_cluster=False, col_cluster=False,figsize=[18,16],center=0.8,)
plt.savefig('heatmap.pdf')
```
